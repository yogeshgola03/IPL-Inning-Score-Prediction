```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset.csv')
```

```python
df
```

# Data Cleaning

```python
# Removing unwanted columns
columns_to_remove = ['Unnamed: 0','mid', 'venue', 'batsman', 'bowler', 'striker', 'non-striker']
df.drop(labels=columns_to_remove, axis=1, inplace=True)
```

```python
df
```

```python
# Keeping only consistent teams
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                    'Mumbai Indians', 'Kings XI Punjab', 'Royal Challengers Bangalore',
                    'Delhi Daredevils', 'Sunrisers Hyderabad']
df = df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
```

```python
df
```

```python
# Removing the first 5 overs data in every match
df = df[df['overs']>=5.0]
```

```python
df
```

```python
# Converting the column 'date' from string into datetime object
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%d-%m-%Y'))
```

```python
df
```

# Data Preprocessing

```python
# Splitting the data into train and test set
#Training data
x_train = df.drop(labels='total', axis=1)[df['date'].dt.year <= 2016]
y_train = df[df['date'].dt.year <= 2016]['total'].values

#Testing data
x_test = df.drop(labels='total', axis=1)[df['date'].dt.year >= 2017]
y_test = df[df['date'].dt.year >= 2017]['total'].values
```

```python
# Removing the 'date' column
x_train.drop(labels='date', axis=True, inplace=True)
x_test.drop(labels='date', axis=True, inplace=True)
```

```python
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

```python
x_train = x_train.values
x_test = x_test.values
```

```python
"""
0 : Kolkata Knight Riders
1 : Chennai Super Kings
2 : Rajasthan Royals
3 : Mumbai Indians
4 : Kings XI Punjab
5 : Royal Challengers Bangalore
6 : Delhi Daredevils
7 : Sunrisers Hyderabad
"""

def team_code(team):
    if team == 'Kolkata Knight Riders':
        return 0
    elif team == 'Chennai Super Kings':
        return 1
    elif team == 'Rajasthan Royals':
        return 2
    elif team == 'Mumbai Indians':
        return 3
    elif team == 'Kings XI Punjab':
        return 4
    elif team == 'Royal Challengers Bangalore':
        return 5
    elif team == 'Delhi Daredevils':
        return 6
    elif team == 'Sunrisers Hyderabad':
        return 7
```

```python
for i in range (37330):
    x_train[i][0] = team_code(x_train[i][0])
    x_train[i][1] = team_code(x_train[i][1])
    if(i<2778):
        x_test[i][0] = team_code(x_test[i][0])
        x_test[i][1] = team_code(x_test[i][1])
        
```
